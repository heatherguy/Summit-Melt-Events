```python
import numpy as np

from bokeh.io import output_notebook, show
from bokeh.layouts import gridplot
from bokeh.palettes import Viridis3
from bokeh.plotting import figure
from bokeh.models import Title

output_notebook()
```

```python
depth   = np.array([-90, -160, -180, -205, -230, -240, -270, -300, -360, -390, -435, -460, -525, -630, -680, -755, -910, -940, -1020])
density = np.array([258.3, 230.6, 252.8, 227.8, 172.2, 152.8, 241.7, 283.3, 291.7, 222.2, 280.6, 366.7, 183.3, 300.0, 327.8, 216.7, 369.4, 244.4, 308.3])
```

```python
p1=figure(plot_width=300,
       plot_height=500,
       x_axis_label='Density (kg m-3)',
       y_axis_label='Depth below surface (cm)',
       y_range=[-120, 0])
p1.line(density, depth/10)
p1.scatter(density, depth/10, color='darkblue', size=8)

p1.add_layout(Title(text="19 October 2019", text_font_style="italic"), 'above')
p1.add_layout(Title(text="Snow pit from Summit Station"), 'above')

show(p1)
```
